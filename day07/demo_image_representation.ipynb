```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading an image
url = 'http://archive.engineering.nyu.edu/files/imagecache/img_col_8_380/pressrelease/RS46906_071A7857-scr.jpg'
img = io.imread(url)
plt.imshow(img)
print(img)
print(img.shape)
```

```python
plt.figure()
plt.imshow(img)

# Red
img_red = img.copy()
img_red[:,:,1] = np.zeros((461, 690))
img_red[:,:,2] = np.zeros((461, 690))
plt.figure()
plt.imshow(img_red)
plt.title('Red')

# Green
img_green = img.copy()
img_green[:,:,0] = np.zeros((461, 690))
img_green[:,:,2] = np.zeros((461, 690))
plt.figure()
plt.imshow(img_green)
plt.title('Green')

# Blue
img_blue = img.copy()
img_blue[:,:,0] = np.zeros((461, 690))
img_blue[:,:,1] = np.zeros((461, 690))
plt.figure()
plt.imshow(img_blue)
plt.title('Blue')
```
