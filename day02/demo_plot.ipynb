```python
import numpy as np
import matplotlib.pyplot as plt
```

## Demos

### scatter() and plot()
- What is the difference between the two functions?

```python
x1 = [0,2,5,4]
y1 = [1,3,2,1]
plt.scatter(x1,y1)

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
plt.plot(x1,y1)
plt.xlabel('x')
plt.xlabel('y')
```

### Straight line
- plot() connects points with a line
- to plot f(x), we need to generate a lot of points (x, f(x))
- use linspace() to generate x coordiantes

```python
x = np.linspace(-1,1,10)
print(x)

y = 2 * x
print (y)
y = 2 * x + 0.5
print (y)
```

- is y a function, a scalar or a vector? 
- what is each element of y?

```python
y = 2*x + 0.5
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
```
